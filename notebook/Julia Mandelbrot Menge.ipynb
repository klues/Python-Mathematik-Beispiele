## Mandelbrot Menge 


![mb](images/mandelbrot.png)

```julia
# Iteration quadratische Zahlen
function mandel_pt(c, MAX_ITER=100)
    iter = 0
    z    = 0
    while (iter < MAX_ITER)
        z = z^2 + c
        iter+= 1
        if abs(z) > 2.0
            return iter
        end
    end
    0
end;
```

## Menge erzeugen

#### Farbcode f√ºr einen Punkt c in der komplexen Ebene


$$
f(z) \mapsto z^2 +c
$$

Start

$$
z_0 = f(0) = c
$$

$$
z_1 = f(c) = c^2 + c
$$

$$
z_2 = f(c^2+c) = (f(c) = c^2 + c)^2
$$


```julia
mandel(zmin, zmax, resolution=300) =
    [ mandel_pt(complex(j, i)) for i=linspace(zmin.im, zmax.im, resolution), j=linspace(zmin.re, zmax.re, resolution) ];
```

```julia
# The complex plane to map
(zmin, zmax) = (-0.5-1.5im, +1.2+1.5im)

# The Mandelbrot set
mandel_set = mandel(zmin, zmax);
```

```julia
using PyPlot;
imshow(mandel_set,cmap="spectral");
```

## Zoom in die Mandelbrot Ebene

```julia
f = figure()
@manipulate for k=0:0.1:6, x=-2:0.01:2, y=-1:0.01:1; withfig(f) do
        z = 2*0.5^k;zmin = x-z+(y-z)*im;zmax = x+z+(y+z)*im
        mandel_set = mandel(zmin, zmax);
        imshow(mandel_set,cmap="spectral");
    end
end
```
