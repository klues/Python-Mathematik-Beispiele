### Graphs with nxGraph 
from Python  

```python
using PyPlot;
using PyCall;
@pyimport networkx as nx
```

```python


G=nx.star_graph(12)
pos=nx.spring_layout(G)
nx.draw(G,pos,node_color="#A0CBE2",width=4)
```

```python
petersen=nx.petersen_graph()
nx.draw(petersen)
```

```python
tutte=nx.tutte_graph()
nx.draw(tutte)
```

```python
maze=nx.sedgewick_maze_graph()
nx.draw(maze)
```
