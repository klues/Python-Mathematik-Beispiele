```python
import numpy as np

N = 200

X = np.random.uniform(-3.,3.,(N,1))
Y = np.sin(X) + np.random.randn(N,1)*0.05
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(X,Y,'*')
```

```python

import GPy

```

```python


kernel = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)

```

```python


m = GPy.models.GPRegression(X,Y,kernel)

```

```python
m
```

```python


fig = m.plot()


```
