```julia
z =BigInt(1)
cells = (Array{Uint8, 1})[]

for i = 1:20
    tmp = bin(z)
    push!(cells,tmp)
    #print(tmp)
    #print("\n")
    z=z*7
end

cells

length(cells)

```

```julia
dim = 100

out=zeros(Int8,dim,dim)
```

```julia
for j = 1:dim
    if length(cells) >= j
        tmp = cells[j]
    else
        tmp = []
    end
    for i = 1:dim
        if length(tmp) >= i
            out[j,dim+1-i]=tmp[i]-48
        else
            out[j,dim+1-i]=0
        end
    end
end
    
out    
    
```

```julia
dim = 500
pot = 500

out=zeros(Int8,dim,dim)

z =BigInt(1)
cells = (Array{Uint8, 1})[]

for i = 1:pot
    tmp = bin(z)
    push!(cells,tmp)
    #print(tmp)
    #print("\n")
    z=z*3
end

cells

length(cells)



for j = 1:dim
    if length(cells) >= j
        tmp = cells[j]
    else
        tmp = []
    end
    for i = 1:dim
        if length(tmp) >= i
            out[j,dim+1-i]=tmp[i]-48
        else
            out[j,dim+1-i]=0
        end
    end
end
    
out  
```

```julia
using PyPlot
using PyCall

title("Potenz")
imgplot = imshow(out)
pycall(imgplot["set_interpolation"],Any,"nearest")
pycall(imgplot["set_cmap"],Any,"hot")
```
