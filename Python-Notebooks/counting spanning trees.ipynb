```python
# Counting Spanning Trees
from math import pi,sin,cos    
def edgeListKGraph(n):
    eL = []
    for j in range(n):
        for i in range(j):
            eL.append([i,j])          
    return eL

def circlePos(n, r=260):
    e=[]
    w = 2*pi/n
    for j in range(n):
        x = cos(j*w+pi/4)*r
        y = sin(j*w+pi/4)*r
        e.append([round(x),round(y)])
    return e

def intToBinA(z):
    s = bin(z)
    a = []
    
    # gleiches lÃ¤nge  fÃ¼r alle
    k = len(edgeL)-len(s)+2
    for i in range(k):
        a.append(0)

    for i in range(2,len(s)):
        a.append(int(s[i]))
    return a

def checkConnectedGraph(eK):
    NL = [0]*N
    for i in range(len(eK)):
        if eK[i] == 1:
            NL[edgeL[i][0]] += 1
            NL[edgeL[i][1]] += 1

    for e in NL:
        if e == 0:
            return False
        
    if anzVerbunden(eK) != N:
        return False
    return True

def anzVerbunden(eK):
    kante=[]
    #Kantengraph aufbauen 
    for i in range(len(eK)):
        if eK[i]==1:
            kante.append([edgeL[i][0],edgeL[i][1]])
    l = kante[0]
    kante.remove(l)
    
    #Verbundene Knoten in Liste einbauen
    i = 0
    while(len(kante)>0 and i < len(kante)):
        k = kante[i]
        if k[0] in l:
            l.append(k[1])
            kante.remove(k)
            i = 0
        else:
            if k[1] in l:
                l.append(k[0])
                kante.remove(k)
                i = 0
            else:
                i = i + 1
    return len(l)
                            

N = 5

ecken = circlePos(N)
edgeL = edgeListKGraph(N)


c = 0
M = 2**len(edgeL)
for j in range(M):
    binA=intToBinA(j)
    if sum(binA)==N-1:
        #print(" ",binA)
        if checkConnectedGraph(binA):   
            c = c + 1
print("   ",c)
print("   ",N**(N-2))
```
