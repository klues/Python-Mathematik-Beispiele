```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 24
```

```python
def sq_N (x,N):
    return x*x % N
```

```python
plt.figure(figsize=(10,10));
x = range(N)

y = [sq_N(i,N) for i in x]

plt.plot(x,y)
```

```python
k = 0
for i in y:
    if i == 1:
        print(i,k-1, N-k-1)
    k = k+1
```
