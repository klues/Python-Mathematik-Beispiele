```r
#Kreieren eines grösseren Datensets mit diesen Mittelwerten und Standardabweichungen:

gr.vec<- c( rnorm(20, mean(W), sd(W) ), rnorm( 20 , mean(M) , sd(M)))
hist(gr.vec)
boxplot( gr.vec ~ wm)
```

```r
#Permutationstest:
# Logik des Tests: Unter der Annahme, dass alle Beobachtungen der 2 Gruppen aus der gleichen Verteilung # stammen (mit gleichem Mittel) - dann sollte jede zufällige Permutation des Datensets gleich
# wahrscheinlich sein, und somit auch die Differenz der zwei Mittelwerte. Wir permutieren nun das Datenset
# N Mal und berechnen dann jedes mal den Absolutbetrag der Differenz. Diese Verteilung wird dann am
# dann gebraucht um zu sehen, wo unsere “beobachtete Differenz” hinfällt. 

Diff.obs<- mean( gr.vec[1:20]) - mean( gr.vec[21:40]); Diff.obs<-abs(Diff.obs); Diff.obs  #gucken absolute Differenz an (2-seitiger test)
```

```r
# zufällige Permutation der Zahlenwerte mit der Funktion sample:

sample(1:100)
sort (sample(1:100))
sort (sample(1:100,100, T)) #zufälliges Ziehen mit zurücklegen

rand <- sample (gr.vec) 
r.diff<- abs ( mean( rand[21:40]) - mean( rand[1:20])  )
r.diff
```

```r
#1000 Zufallsdifferenzen in einem Vektor speichern; leeren Vektor generieren:

rDiff.1000 <- vector()

for (i in 1:1000) {
rand <- sample (gr.vec) 
r.diff<-abs( mean( rand[21:40]) - mean( rand[1:20]) )
rDiff.1000[i]<-r.diff
}
hist(rDiff.1000)
```

```r
rDiff.1000[1001] <- Diff.obs

 p.val<- length( rDiff.1000[ rDiff.1000 >= Diff.obs]) / 1001

p.val
```

```r
#cleans out the workspace:  rm(list=(ls()))

#als Funktion in einem go:

rm(list=(ls()))

W<-c(165,174,173)
M<-c(172,185,177,181,175)

Diff.perm.test<-function(group.size, N.perm){

dat <- c(rnorm( group.size , mean(W),sd(W)), rnorm( group.size , mean(M),sd(M)))   


diff.o <-abs( mean(dat[1:group.size]) - mean(dat[(group.size+1):(2*group.size)]) )

randD<- vector()
    
for (i in 1: N.perm ) {
perm.vec <- sample (dat)                                                                      #permutierter Datenvektor 
randD[i] <- abs( mean( perm.vec[1:20]) - mean( perm.vec[21:40]) ) # Absolutbetrag der Differenz wird im Vektor rDiff an der Stelle i abgespeichert
}
randD[N.perm+1]<-diff.o             #der beobachtete Wert wird auch dazugenommen

p.val<- length( randD[ randD >= diff.o] ) / (N.perm+1)

print("Die beobachtete Differenz ist"); print(diff.o) 
print( "Der p-Wert ist"); print(p.val)  

}

Diff.perm.test(30,1000)

```
