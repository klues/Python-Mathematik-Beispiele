```r
library(GenSA)

Rastrigin <- local({ 
  index <- 0  
  function(x){    
    index <<- index + 1    
    if(index%%1000 == 0){
      cat(index, "   ")
    }    
    sum(x^2 - 10*cos(2*pi*x)) + 10*length(x)    
  }  
})

set.seed(1234)
dimension <- 10
lower <- rep(-5.12, dimension)
upper <- rep(5.12, dimension)
out <- GenSA(lower = lower, upper = upper, fn = Rastrigin, control = list(max.call = 10^4))

```

```r
f2 <- function(x){
    1*x*x*x-2*x*x+3*x-4
}


reg <- function(p){
    ft <- function(x){
        p[1]*x*x*x+p[2]*x*x+p[3]*x+p[4]
    }
    
    start <- -10
    end <- 10
    

    pos <- start
    step <- 0.1
    res <-0
    while(pos < end) {
        res = res + sqrt((ft(pos)-f2(pos))^2)
        pos = pos + step
    }
    steps <- (end-start)/step
    return(res/steps)
    
    
}
```

```r
v = c(1,-2,3,-4)

print(reg(v))
```

```r
set.seed(14)
dimension <- 4
lower <- rep(-5, dimension)
upper <- rep(5, dimension)
out <- GenSA(lower = lower, upper = upper, fn = reg, control = list(max.call = 10^4))
out
```
